```python
import numpy as np
import matplotlib.pyplot as plt

# 定义角度范围
theta = np.linspace(0, 2*np.pi, 1000)

# 计算半径
r = 10* np.cos(2*(theta - np.pi/2) )

# 绘制极坐标图表
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rmax(max(r)) # 设置极径最大值为 r 的最大值
ax.set_rticks([-10,10]) # 设置极径刻度
ax.set_rlabel_position(-22.5) # 设置极径标签位置
ax.grid(False) # 添加网格线
plt.show()
```
